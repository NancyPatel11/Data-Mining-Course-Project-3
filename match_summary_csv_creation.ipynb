```python
df_matches = df_matches.rename(columns={'match_number': 'match_id'})
merged_df = pd.merge(df_deliveries, df_matches, on='match_id')
```

```python
merged_df.columns
```

```python
selected_features=['match_id', 'venue_x', 'innings', 'batting_team', 'bowling_team',
                   'ball', 'runs_off_bat', 'wides', 'noballs', 'byes', 'legbyes', 'extras']
```

```python
merged_df=merged_df[selected_features]
```

```python
merged_df[["wides", "noballs", "extras"]] = merged_df[["wides", "noballs", "extras"]].fillna(0)
```

```python
merged_df["total_runs_on_this_ball"] = merged_df["runs_off_bat"] + merged_df["extras"]
```

```python
total_runs_per_inning_match = merged_df.groupby(['match_id', 'innings'])['total_runs_on_this_ball'].sum().reset_index(name="total_runs_per_inning_match")
merged_df = merged_df.merge(total_runs_per_inning_match, on=['match_id', 'innings'], how='left')
```

```python
b1 = merged_df.groupby(['match_id', 'innings'])['ball'].size().reset_index(name ='total_balls_including_wides_noball')
b2 = merged_df.groupby(['match_id', 'innings'])['wides'].apply(lambda x: (x != 0).sum()).reset_index(name="total_wides")
b3 = merged_df.groupby(['match_id', 'innings'])['noballs'].apply(lambda x: (x != 0).sum()).reset_index(name='total_noballs')
```

```python
merged_df = merged_df.merge(b1, on=['match_id', 'innings'], how='left')
merged_df = merged_df.merge(b2, on=['match_id', 'innings'], how='left')
merged_df = merged_df.merge(b3, on=['match_id', 'innings'], how='left')
```

```python
merged_df["Total_Overs_Played"] = (merged_df["total_balls_including_wides_noball"] - merged_df["total_wides"] - merged_df["total_noballs"])/6
```
